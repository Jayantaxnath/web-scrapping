```python
import pandas as pd
import numpy as np
import glob
import os
from uuid import uuid4

print("Libraries imported successfully!")
```

```python
# List all CSV files that contain internship data
csv_files = [
    'internship-in-bangalore.csv',
    'internship-in-chennai.csv',
    'internship-in-delhi.csv',
    'internship-in-hyderabad.csv',
    'internship-in-jaipur.csv',
    'internship-in-kolkata.csv',
    'internship-in-mumbai.csv',
    'internship-in-pune.csv',
    'work-from-home-internships.csv'
]

print(f"Found {len(csv_files)} CSV files to merge:")
for file in csv_files:
    if os.path.exists(file):
        print(f"✓ {file}")
    else:
        print(f"✗ {file} - File not found")
```

```python
# Read and merge all CSV files
dataframes = []
total_records = 0

for file in csv_files:
    if os.path.exists(file):
        try:
            df = pd.read_csv(file)
            print(f"Loaded {file}: {len(df)} records")
            total_records += len(df)
            dataframes.append(df)
        except Exception as e:
            print(f"Error reading {file}: {e}")
    else:
        print(f"File {file} not found, skipping...")

# Combine all dataframes
if dataframes:
    merged_df = pd.concat(dataframes, ignore_index=True)
    print(f"\nSuccessfully merged {len(dataframes)} files")
    print(f"Total records: {total_records}")
    print(f"Final dataframe shape: {merged_df.shape}")
    print(f"Columns: {list(merged_df.columns)}")
else:
    print("No dataframes to merge!")
    merged_df = pd.DataFrame()
```

```python
# Shuffle the merged dataframe
if not merged_df.empty:
    print("Shuffling the merged data...")
    shuffled_df = merged_df.sample(frac=1, random_state=42).reset_index(drop=True)
    print("Data shuffled successfully!")
    
    # Display first few rows to verify
    print("\nFirst 5 rows after shuffling:")
    print(shuffled_df.head())
else:
    print("No data to shuffle!")
    shuffled_df = pd.DataFrame()
```

```python
# Add unique internship IDs
if not shuffled_df.empty:
    print("Adding unique internship IDs...")
    
    # Generate unique IDs with format: 2025WSHP0001, 2025WSHP0002, etc.
    # Format: Year(2025) + WS(Web Scraping) + HP(User initials) + 4-digit sequential number
    internship_ids = [f"2025WSHP{str(i+1).zfill(4)}" for i in range(len(shuffled_df))]
    
    # Insert the ID column at the beginning
    final_df = shuffled_df.copy()
    final_df.insert(0, 'internship_id', internship_ids)
    
    print(f"Added {len(internship_ids)} unique internship IDs")
    print(f"ID format: 2025WSHP0001 to 2025WSHP{str(len(internship_ids)).zfill(4)}")
    
    # Display sample with IDs
    print("\nSample data with internship IDs:")
    print(final_df[['internship_id', 'profile', 'company', 'Location']].head(10))
else:
    print("No data to add IDs to!")
    final_df = pd.DataFrame()
```

```python
# Save the final processed dataset
if not final_df.empty:
    output_filename = 'processed_internships_with_ids.csv'
    final_df.to_csv(output_filename, index=False)
    
    print(f"Dataset saved successfully to '{output_filename}'")
    print(f"Total records: {len(final_df)}")
    print(f"File size: {os.path.getsize(output_filename) / (1024*1024):.2f} MB")
    
    # Display final dataset info
    print(f"\nFinal dataset summary:")
    print(f"Shape: {final_df.shape}")
    print(f"Columns: {list(final_df.columns)}")
    print(f"Memory usage: {final_df.memory_usage(deep=True).sum() / (1024*1024):.2f} MB")
    
    print("\nProcessing completed successfully! ✅")
    print(f"✓ Merged 9 CSV files")
    print(f"✓ Shuffled {len(final_df)} records") 
    print(f"✓ Added unique IDs (2025WSHP0001 to 2025WSHP{str(len(final_df)).zfill(4)})")
    print(f"✓ Saved to '{output_filename}'")
else:
    print("No data to save!")
```
